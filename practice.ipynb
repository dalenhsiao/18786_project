```python
from torch_geometric.nn import GCNConv, GAE
class Encoder(nn.Module):
    def __init__(self, in_channel, hidden_channel, output_channel):
        """
        node features: (n_node, F_in)
        edge indices: (2, num_edges) or sparse matrix 
        """
        super(Encoder, self).__init__()
        self.in_channel = in_channel
        self.hidden_channel = hidden_channel
        self.output_channel = output_channel
        self.embedding = nn.Linear(in_channel, hidden_channel)

        self.conv1 = GCNConv(in_channel, hidden_channel) 

    def forward(self, x):
        h = torch.relu(self.fc1(x))
        h = torch.relu(self.fc2(h))
        return self.fc3(h)
```
